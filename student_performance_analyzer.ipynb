```python
import numpy as np
```

```python
data = np.loadtxt('./students.csv',delimiter=",", skiprows=1)
print(data)
```

```python
maths = data[:,1]
physics = data[:,2]
chemistry = data[:,3]
print(maths,physics,chemistry)
```

```python
print(np.mean(maths),np.mean(physics),np.mean(chemistry))
```

```python
Topper_student_marks = np.max(np.sum(data,axis=1))
print(Topper_student_marks)
```

```python
avg_marks = np.mean(data[:,1:],axis=1)
print(avg_marks)
```

```python
Grades = np.array(["","","","",""])
Grades[avg_marks>=85]="A"
Grades[(avg_marks>=70) & (avg_marks<85)]="B"
Grades[(avg_marks>=55) & (avg_marks<70)]="C"
Grades[avg_marks<55]="D"
```

```python
print(Grades)
```
